```python
#Data Frame 2- Country Code
df_country = pd.read_excel('/content/drive/MyDrive/ColabNotebooks/Zomato_EDA/Dataset/Country-Code.xlsx')
```

```python
df_country.head()
```

```python
#Combining the 2 Data Frames on same feature country code
final_df = pd.merge(df, df_country, on='Country Code', how = 'left')
final_df.head()
```

```python
final_df.Country.value_counts()
```

```python
country_name = final_df.Country.value_counts().index
```

```python
country_value_cnt = final_df.Country.value_counts().values
```

```python
#Top 3 countries using Zomato
figure(figsize=(8, 6), dpi=80)
plt.pie(country_value_cnt[:3], labels= country_name[:3], autopct="%1.2f%%", )
plt.show()
```

Maximum records are from India. Followed by the USA and the UK.

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
figure(figsize=(10, 5), dpi=80)
plt.rcParams['axes.facecolor'] = 'gray'
sns.barplot(data = ratings, x = 'Aggregate rating', y = 'Rating count', hue = 'Rating color',
            palette = ['white', 'red', 'orange', 'yellow', '#90EE90', 'green'], dodge = False)
```

**Data insights.** 

More than 2000 people have not rated the Zomato.

Many people gave the ratings between 2.6 to 3.9 which is average rating.

Very less pople gave rating between 1.0 to 2.4 and 4.7 to 4.9.

```python
#rating count by Country
country_ratings = final_df.groupby(['Country', 'Aggregate rating']).size().reset_index()
country_ratings
```

```python
#Cuisines with Aggregate rating greater than 4.0
ratings_cuisine = final_df[final_df['Aggregate rating']>=4.0].groupby(['Cuisines', 'Aggregate rating']).size().reset_index().rename(columns={0:'Rating count'})
ratings_cuisine
```

```python
## Top Indian cities with highest orders

orders_ind_cnt = final_df[final_df['Country'] == 'India']
top_cities_ind = orders_ind_cnt['City'].value_counts()
top_cities_ind[:10]
```

```python
top_cities_ind[:10].plot(kind= 'bar', )
plt.xlabel('City')
plt.ylabel('Orders')
```

Insights: Top Indian cities with highest records

New Delhi has the highest number of orders. Followed by Gurgaon, Noida and Faridabad.
